```python
decibel_values=librosa.power_to_db(power_values)
### frequencies list
frequencies=librosa.fft_frequencies(sr=sr,n_fft=2048)
max_intensity_freqs=np.array([])

for i in range(0,decibel_values.shape[1]):
           max_value=np.argmax(decibel_values[:,i])
           max_intensity_freqs=np.append(max_intensity_freqs,frequencies[max_value])
    
```

```python
### generating time values for each sample of stft window
time_values=np.array([])
time_samples=librosa.get_duration(y)/decibel_values.shape[1]
temp=0
for i in range(0,decibel_values.shape[1]):
        temp+=time_samples
        time_values=np.append(time_values,temp)
```

```python
new_freqs=np.array([])
new_time_values=np.array([])
temp_value=0
for i in range(0,max_intensity_freqs.size):
    if temp_value!=max_intensity_freqs[i]:
            new_freqs=np.append(new_freqs,max_intensity_freqs[i])
            new_time_values=np.append(new_time_values,time_values[i])
            temp_value=max_intensity_freqs[i]



                
```

```python
freqs_data=np.array([])
new_freqs=new_freqs.astype(int) # converting to int
hashes_list=np.array([]) ### hash values of frequencies plus offset
time_offsets=np.array([]) ### time offsets
target_zone=5 #five forward frequencies
time_offest_divison=1000 ##offset 4 decimal wants
for i in range(0,new_freqs.size):
    for j in range(1,target_zone+1):
        if(i+j)<new_freqs.size:
            freq_temp=int(str(new_freqs[i])+str(new_freqs[i+j]))
            time_temp=int((new_time_values[i+j]-new_time_values[i])*1000)
            time_offsets=np.append(time_offsets,abs(time_temp))
            hashes_list=np.append(hashes_list,str(freq_temp)+str(time_temp))

```

```python
##pattern matching
default_values= np.array(['8188692', '81875743', '81864766'])
possibles=np.where(hashes_list==default_values[0])
solution=np.array([])

for i in possibles:
        check = hashes_list[i:(i+default_values.size)]
        if np.all(check == default_values):
                solution=np.append(solution,i)
        
solution
```
